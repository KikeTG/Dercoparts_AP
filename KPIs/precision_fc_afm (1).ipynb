```python

fecha_actual = datetime.today()

# Calcular la fecha del mes anterior
fecha_mes_anterior = fecha_actual - timedelta(days=fecha_actual.day)

# Formatear la fecha en "yyyy-mm"
fecha_formateada = fecha_mes_anterior.strftime('%Y-%m')
dtype_stock = {
    'Material SAP': str,
    'UE': str,
}

# Obtener el valor del diccionario para el mes anterior al actual
semanas_habiles_mes_anterior = semanas_habiles_por_mes.get(fecha_formateada, 4)

# Ajustar el rango del bucle for según el valor del diccionario
if semanas_habiles_mes_anterior == 4:
    rango_bucle = 6
elif semanas_habiles_mes_anterior == 5:
    rango_bucle = 7
else:
    # Ajusta según tus necesidades si el valor no es ni 4 ni 5
    rango_bucle = 6

tubo = os.listdir(carpeta_tubo)
# Iterar sobre las carpetas y archivos
for i in tubo:
    
    if nombre_ciclo_n1 in i:
        
        
        arch = os.path.join(carpeta_tubo, i)
        print(arch)
        df = pd.read_excel(arch, sheet_name='Sheet1', usecols=['UE','Venta UMB', 'Libre utilización'])
        df['UE'] = df['UE'].astype(str)
        
        
        
        df_ue = pd.merge(df, cadena_de_remplazo, left_on='UE', right_on='Nro_pieza_fabricante_1', how='left', suffixes=('UE', '_Cad_reemplazo'))
        
        df_ue.drop('Nro_pieza_fabricante_1', axis=1, inplace=True)
        
        df_ue['Cod_Actual_1'] = df_ue['Cod_Actual_1'].fillna(df_ue['UE'])

        
        def calcular_instock(row):
            if row['Venta UMB'] == 0 and row['Libre utilización'] == 0:
                return 0
            elif row['Venta UMB'] > row['Libre utilización']:
                return 0
            else:
                return 1

# Aplicar la función a cada fila del DataFrame
        df_ue['Instock'] = df.apply(calcular_instock, axis=1)
        df_ue = df_ue.groupby('Cod_Actual_1').agg({'Instock': 'first'}).reset_index()
        
        df_ue = df_ue.rename(columns = {'Instock': f'Instock SEM {i[-7:-5]}', 'Cod_Actual_1':f'Cod_Actual{i[-7:-5]}'})
  

        suffix = f'_{i}' if i == (len(tubo) - 2) else f'_{i}a'  # Añadir sufijo para la primera fusión
        consolidado_cruce_ventas = consolidado_cruce_ventas.merge(df_ue, left_on='UE_2', right_on=f'Cod_Actual{i[-7:-5]}', how='left')
```
