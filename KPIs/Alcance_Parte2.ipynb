```python
import os
import pandas as pd

# Función para obtener el archivo parquet más reciente
def obtener_archivo_reciente(ruta):
    archivos = [os.path.join(ruta, archivo) for archivo in os.listdir(ruta) if archivo.endswith('.parquet')]
    if archivos:
        archivo_reciente = max(archivos, key=os.path.getctime)
        return archivo_reciente
    else:
        raise FileNotFoundError(f"No se encontraron archivos .parquet en la ruta: {ruta}")

# Obtener el usuario actual del sistema
usuario = os.path.expanduser('~').split('\\')[-1]

# Rutas de las carpetas (adaptadas al usuario actual)
rutas = [
    fr"C:\Users\{usuario}\DERCO CHILE REPUESTOS SpA\Planificación y abastecimiento - Documentos\KPI\Gerenciamiento Alcances\PBI\TR\Consolidados\FC_Promedio",
    fr"C:\Users\{usuario}\DERCO CHILE REPUESTOS SpA\Planificación y abastecimiento - Documentos\KPI\Gerenciamiento Alcances\PBI\TR\Consolidados\Stock",
    fr"C:\Users\{usuario}\DERCO CHILE REPUESTOS SpA\Planificación y abastecimiento - Documentos\KPI\Gerenciamiento Alcances\PBI\TR\Consolidados\Transito",
    fr"C:\Users\{usuario}\DERCO CHILE REPUESTOS SpA\Planificación y abastecimiento - Documentos\KPI\Gerenciamiento Alcances\PBI\TR\Consolidados\Venta_Promedio",
    fr"C:\Users\{usuario}\DERCO CHILE REPUESTOS SpA\Planificación y abastecimiento - Documentos\KPI\Gerenciamiento Alcances\PBI\Stock_tienda\sin_centro"
]

# Cargar los archivos parquet más recientes
dfs = []
for ruta in rutas:
    try:
        archivo_reciente = obtener_archivo_reciente(ruta)
        df = pd.read_parquet(archivo_reciente)
        dfs.append(df)
    except FileNotFoundError as e:
        print(e)
        dfs.append(None)

# Asignar a variables por cada DataFrame
df_fc_promedio, df_stock, df_transito, df_venta_promedio, df_stock_tienda = dfs

# Comprobación de las cargas
for nombre, df in zip(['df_fc_promedio', 'df_stock', 'df_transito', 'df_venta_promedio', 'df_stock_tienda'], dfs):
    if df is not None:
        print(f"{nombre} cargado con éxito. {len(df)} filas.")
    else:
        print(f"{nombre} no pudo ser cargado.")
```
