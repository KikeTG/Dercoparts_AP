```python
import pandas as pd
import os
from datetime import datetime

# Definir los directorios de entrada y salida
input_dir = r"C:\Users\etorres.DERCOPARTS\DERCO CHILE REPUESTOS SpA\Planificación y abastecimiento - Documentos\KPI\Tubo Semanal\2024-08-19\monitor"
output_dir = r"C:\Users\etorres.DERCOPARTS\DERCO CHILE REPUESTOS SpA\Planificación y abastecimiento - Documentos\KPI\Data Science\OC_PROD\Monitor"

# Obtener la fecha de hoy
fecha_hoy = datetime.now().strftime('%Y-%m-%d')

# Crear una lista para almacenar los DataFrames
df_list = []

# Leer todos los archivos Excel en la carpeta especificada
for file in os.listdir(input_dir):
    if file.endswith('.XLSX'):
        file_path = os.path.join(input_dir, file)
        # Leer el archivo Excel forzando todas las columnas como cadenas
        df = pd.read_excel(file_path, dtype=str)
        df_list.append(df)

# Concatenar todos los DataFrames
df_concatenado = pd.concat(df_list, ignore_index=True)



# Definir el nombre del archivo de salida
output_file = os.path.join(output_dir, f'concatenado_{fecha_hoy}.parquet')

# Guardar el DataFrame concatenado como un archivo Parquet
df_concatenado.to_parquet(output_file)
```
