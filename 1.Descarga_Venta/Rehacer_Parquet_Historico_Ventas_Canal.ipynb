```python
df_melted = pd.melt(
    df,
    id_vars=["Canal", "Material", "id"],
    var_name="Variable",
    value_name="Valor"
)
df_melted
```

```python
total_valor = df_melted.loc[df_melted["Variable"] == "2024-04-01 00:00:00", "Valor"].sum()
total_valor
```

```python
df_melted = df_melted.rename(columns={"Variable": "Mes", "Valor": "Venta UMB"})
df_melted["Canal 3"] = df_melted["Canal"].apply(
    lambda x: "CL AGROPLANET" if x == "Agroplanet" else "CL AUTOPLANET"
)
```

```python
df3 = pd.concat([df2, df_melted], ignore_index=True)
```

```python
df3["Material"] = df3["Material"].astype(str)

output_path = r"C:\Users\etorres.DERCOPARTS\DERCO CHILE REPUESTOS SpA\Planificación y abastecimiento - Documentos\Planificación y Compras AFM\S&OP Demanda\Codigos Demanda\Parquets\Historia_Sell_Out (2024-10)_Canales.parquet"
df3.to_parquet(output_path, index=False)
```

```python
# Crear el pivot table con valores vacíos reemplazados por 0
pivot_df3 = df3.pivot_table(
    index=["Material", "Canal"],
    columns="Mes",
    values="Venta UMB",
    aggfunc="sum",
    fill_value=0
).reset_index()

# Guardar el archivo como Excel
output_excel_path = r"C:\Users\etorres.DERCOPARTS\DERCO CHILE REPUESTOS SpA\Planificación y abastecimiento - Documentos\Planificación y Compras AFM\S&OP Demanda\Codigos Demanda\Parquets\Historia_Sell_Out (2024-10)_Canales_pivot.xlsx"
pivot_df3.to_excel(output_excel_path, index=False)
```
