## LEER HISTORIA DE VENTA

```python
import pandas as pd
file_path=r"C:\Users\etorres.DERCOPARTS\DERCO CHILE REPUESTOS SpA\Planificación y abastecimiento - Documentos\Planificación y Compras Anastasia\Carga Historia de Venta\2024-10 Ciclo Nov-24\AFM\Correccion Dispo\Anastasia 2024.10.csv"
df_anastasia=pd.read_csv(file_path, delimiter=',') 

df_anastasia['Fecha'] = pd.to_datetime(df_anastasia['Fecha']) 
df_anastasia = df_anastasia[df_anastasia['Fecha'] < '2024-09-01']
df_anastasia = df_anastasia[df_anastasia['Canal'] == 'CL RETAIL']
df_anastasia = df_anastasia[['Fecha','idSKU','Cantidad']].rename(columns={'idSKU':'sku'})

# Obtener las 24 fechas únicas más recientes
recent_dates = df_anastasia['Fecha'].drop_duplicates().nlargest(24)

# Filtrar las filas que contienen una de las 24 fechas únicas más recientes
df_anastasia = df_anastasia[df_anastasia['Fecha'].isin(recent_dates)]
```

```python
a=df_anastasia['Fecha'].unique()
a
```

```python
td_vtas =df_anastasia.pivot(index='sku', columns='Fecha', values='Cantidad').fillna(0).reset_index()
```

```python
td_vtas.head()
```

```python
initial_rows_td_vtas = len(td_vtas)
initial_rows_prices = len(prices)
removed_rows_td_vtas = initial_rows_td_vtas - len(td_vtas)
removed_rows_prices = initial_rows_prices - len(prices)

print(f"Filas eliminadas de td_vtas: {removed_rows_td_vtas}")
print(f"Filas eliminadas de prices: {removed_rows_prices}")

for i in range(2, 25):
    col_name = f'v_{i-1}'
    td_vtas.loc[:, col_name] = (td_vtas.iloc[:, i] - td_vtas.iloc[:, i-1]) / td_vtas.iloc[:, i-1]
    td_vtas.loc[:, col_name] = td_vtas[col_name].fillna(0)

for i in range(2, 25):
    col_name = f'p_{i-1}'
    prices.loc[:, col_name] = (prices.iloc[:, i] - prices.iloc[:, i-1]) / prices.iloc[:, i-1]
    prices.loc[:, col_name] = prices[col_name].fillna(0)
```
